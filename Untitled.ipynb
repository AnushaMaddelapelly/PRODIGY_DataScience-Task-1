```python
df=pd.read_csv('Task 1.csv',index_col=0)
df.index=df.index.str.strip()
if 'Population, female' in df.index:
    female_population=df.loc["Population, female"]
    print("Population, female Data:\n",female_population)
else:
    print("Error: 'Population, female' not found in the index.")
    print("Available indices:",df.index.tolist())
```

```python
data={
    'year':[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023],
    'Female Population':[599047591,607376801,615647779,623809180,631697152,639323292,647012921,654607791,661854076,668786993,675389679,681060412,685992675,691776131]
}
df=pd.DataFrame(data)
plt.bar(df['year'], df['Female Population'])
plt.xlabel('year')
plt.ylabel('population female')
plt.title('Female Population in India from 2010-2023')
plt.show()
```

```python
data={
    'country_names':['canada','Bahrain','Australia','India'],
    'total_population':[373790876,13150775,250360863,1453404513]
}
df=pd.DataFrame(data)
plt.figure(figsize=(8,6))
plt.bar(df['country_names'],df['total_population'],color='red')
plt.xlabel('country')
plt.ylabel('Total Population')
plt.title('Population total from 2014 to 2023')
plt.show()
```
